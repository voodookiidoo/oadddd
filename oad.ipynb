```python
class WeightGraph:

	def __init__(self, nodes_data: list, nodes_weights: dict):
		self.__nodes = nodes_data  # список вершин графа
		self.__size = len(nodes_data)  # количество вершин в графе
		self.__weights = self.__get_weights(nodes_weights)  # словарь, ключ - ребро в виде фрозенсета с двумя вершинами,
		# значение - вес ребра
		self.__distances = nodes_weights  # словарь, ключ - вершина, значение - список списков [вершина2, вес ребра до неё]
		self.__data = self.__get_data(nodes_weights)  # словарь, ключ - вершина, значение - список доступных вершин
		self.__complete = self.check_completeness()

	def check_completeness(self) -> bool:
		for node in self.__nodes:
			if len(self.__data[node]) < self.__size - 1:
				return False
		return True

	@property
	def nodes(self):
		return self.__nodes

	@property
	def complete(self):
		return self.__complete

	@property
	def nodes_data(self):
		return self.__nodes

	@property
	def weights_dict(self):
		return self.__weights

	@property
	def data(self):
		return self.__data

	def build_voyajer_cycle(self, starting_node: str):
		if not self.__complete:
			raise Exception
		current_node = starting_node
		visited = set()
		frontier = [node for node in self.__data[starting_node] if node not in visited]
		path = [starting_node]
		visited.add(starting_node)
		total = 0
		while len(visited) != self.__size:
			if not frontier:
				break
			prev = current_node
			current_node = min(frontier, key=lambda x: self.__weights[frozenset([current_node, x])])
			total += self.__weights[frozenset([prev, current_node])]
			visited.add(current_node)
			path.append(current_node)
			frontier = [node for node in self.__data[current_node] if node not in visited]
		total += self.__weights[frozenset([starting_node, current_node])]
		path.append(starting_node)
		return path, total

	@staticmethod
	def __get_weights(weights: dict) -> dict:
		weights_dict = dict()
		for first_node, node_weights in weights.items():
			for second_node, distance in node_weights:
				pair = frozenset([first_node, second_node])
				if pair not in weights_dict.keys():
					weights_dict.update({pair: distance})
		return weights_dict

	def __get_data(self, weights):
		result = dict()
		for node in self.__nodes:
			result[node] = list()
			for second in weights[node]:
				result[node].append(second[0])
		return result

	def __getitem__(self, key: frozenset) -> int:
		if key not in self.__weights.keys():
			raise KeyError
		return self.__weights[key]


nodes = 'A B C D'.split()
weights = {'A': [['B', 5], ['C', 6], ['D', 8]], 'B': [['A', 5], ['C', 7], ['D', 10]],
           'C': [['A', 6], ['B', 7], ['D', 3]], 'D': [['A', 8], ['B', 10], ['C', 3]]}
graph = WeightGraph(nodes, weights)

for check_node in graph.nodes:
	print('Start from {}, path is {}, weight is {}\n'.format(check_node, *graph.build_voyajer_cycle(check_node)))
```
